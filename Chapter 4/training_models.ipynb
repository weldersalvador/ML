Linear Regression

Closed form: normal equation

```python
import numpy as np

np.random.seed(42)

m = 100
X = 2 * np.random.rand(m,1)
y = 4 + 3 * X * np.random.rand(m,1)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X,y,marker = '.')
plt.axis([0,2,0,14])
plt.grid()
plt.show()
```

```python
from sklearn.preprocessing import add_dummy_feature

X_b = add_dummy_feature(X)
2
best_theta = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y
```

```python
best_theta
```

```python
f = lambda x : best_theta[0] + best_theta[1]*x

plt.plot(X, f(X),color = 'red')
plt.scatter(X,y,marker = '.',color = 'black')
plt.axis([0,2,0,14])
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear regression")
plt.grid()
plt.savefig("Linear regression (Closed form)")
plt.show()
```

Gradient Descent

```python
learning_rate = 0.1

n_epochs = 100

m = len(X_b)

np.random.seed(42)

theta = np.random.randn(2,1)
```

```python
plt.scatter(X,y,marker = '.',color = 'black')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear regression with batch gradient descent")
plt.grid()
for epoch in range (n_epochs):
    gradients = (2/m) * X_b.T @ (X_b @ theta - y)
    theta = theta - learning_rate * gradients
    f = lambda x : theta[0] + x*theta[1]
    plt.plot(X,f(X),color = 'blue')
plt.savefig("LR with batch gd")
plt.show()
    
```
