```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
data = "https://github.com/ageron/data/raw/main/lifesat/lifesat.csv"
dataFrame = pd.read_csv(data)
```

```python
dataFrame
```

```python
X = dataFrame[["GDP per capita (USD)"]]
Y = dataFrame[["Life satisfaction"]]
```

```python
model = LinearRegression()
model.fit(X,Y)
```

```python
model.intercept_
```

```python
model.predict([[50000]])
```

```python
coef = model.coef_
func = coef*X + model.intercept_
```

```python
dataFrame.plot(kind = 'scatter', x = "GDP per capita (USD)", y = "Life satisfaction")
plt.plot(X,func)
plt.show()
```
